# 一、PaddleHub简介

飞桨预训练模型应用工具PaddleHub旨在为开发者提供丰富的、高质量的、直接可用的预训练模型。通过十行代码轻松完成所有主流人工智能算法的应用，例如目标检测、人脸识别、语义分割以及文本生成等任务。此外，PaddleHub提供了Fine-tune API，开发者可以使用高质量的预训练模型结合Fine-tune API快速完成模型迁移到部署的全流程。

* 模型种类丰富：
涵盖大模型、计算机视觉、自然语言处理、语音、视频等工业应用主流的预训练模型，且全部开源，离线可运行；
* 超低使用门槛：
没有深度学习技术背景，没有数据与训练过程即可使用AI模型；
* 一键模型快速预测：
通过一行命令行或者极简的Python API实现模型调用，可快速体验模型效果；
* 一键模型转服务化：
一行命令，搭建深度学习模型API服务化部署能力；
* 跨平台兼容性：
可运行于Linux、Windows和MacOS等多种操作系统。


PaddleHub涵盖了多个领域的模型：

`ernie_gen_lover_words`通过加载ERNIE 1.0参数热启动，采用网络搜集的数据微调，可用于生成指定主题的文本。

### 行人检测

行人检测是计算机视觉技术中的目标检测问题，用于判断图像中是否存在行人并给予精确定位，定位结果用矩形框表示。行人检测技术有很强的使用价值，它可以与行人跟踪、行人重识别等技术结合，应用于汽车无人驾驶系统、智能视频监控、人体行为分析、客流统计系统、智能交通等领域。

```python
import paddlehub as hub
import cv2
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
import os

pedestrian_detector = hub.Module(name="yolov3_darknet53_pedestrian")

output_path = '/home/aistudio/work/output'

result = pedestrian_detector.object_detection(images=[cv2.imread('/home/aistudio/work/pedestrian.jpg')],
            output_dir=output_path,
            visualization=True)

img = mpimg.imread(os.path.join(output_path,"image_numpy_0.jpg"))

plt.figure(figsize=(15,15))
plt.imshow(img) 
plt.axis('off') 
plt.show()
```

## 2.3 PaddleHub在自然语言处理领域的实践


### 中文情感倾向分析
情感倾向分析是指对带有主观描述的中文文本，自动判断该文本的情感极性类别并给出相应的置信度。通过对商品或服务评价的分析，能够帮助企业理解用户消费习惯，为企业提供有利的决策支持。

```python
import paddlehub as hub

senta = hub.Module(name="senta_bilstm")
test_text = ["这家餐厅很好吃", "这部电影真的很差劲"]
results = senta.sentiment_classify(texts=test_text, 
                                   use_gpu=False,
                                   batch_size=1)
    
for result in results:
    print(result['text'])
    print(result['sentiment_label'])
    print(result['sentiment_key'])
    print(result['positive_probs'])
    print(result['negative_probs'])
```

### 快递单识别

通过AI的方式识别快递单，为物流提供便捷方案。

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
import paddlehub as hub
import cv2

# 待预测图片
test_img_path = [ "/home/aistudio/work/express.jpg"]

# 加载移动端预训练模型
ocr = hub.Module(name="chinese_ocr_db_crnn_mobile")

# 读取测试文件夹test.txt中的照片路径
np_images =[cv2.imread(image_path) for image_path in test_img_path] 

results = ocr.recognize_text(
                    images=np_images,         # 图片数据，ndarray.shape 为 [H, W, C]，BGR格式；
                    use_gpu=False,            # 是否使用 GPU；若使用GPU，请先设置CUDA_VISIBLE_DEVICES环境变量
                    output_dir='ocr_result',  # 图片的保存路径，默认设为 ocr_result；
                    visualization=True,       # 是否将识别结果保存为图片文件；
                    box_thresh=0.5,           # 检测文本框置信度的阈值；
                    text_thresh=0.5)          # 识别中文文本置信度的阈值；

for result in results:
    data = result['data']
    save_path = result['save_path']
    for infomation in data:
        print('text: ', infomation['text'], '\nconfidence: ', infomation['confidence'], '\ntext_box_position: ', infomation['text_box_position'])

# 展示其中广告信息图片
img1 = mpimg.imread(test_img_path[0]) 
plt.figure(figsize=(10,10))
plt.imshow(img1) 
plt.axis('off') 
plt.show()
```
